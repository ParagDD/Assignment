Insights to be drawn -

● Get all Pokemons whose spawn rate is less than 5%

● Get all Pokemons that have less than 4 weaknesses

● Get all Pokemons that have no multipliers at all

● Get all Pokemons that do not have more than 2 evolutions

● Get all Pokemons whose spawn time is less than 300 seconds.

Note - spawn time format is "05:32”, so assume “minute: second” format and perform the analysis.

● Get all Pokemon who have more than two types of capabilities



```python
import pandas as pd
import matplotlib.pyplot as plt


def read_data_from_excel(filename):
    df = pd.read_excel(filename)
    return df.to_dict('records')


def draw_bar_plot(data, xlabel, ylabel, title):
    labels = [pokemon['name'] for pokemon in data]
    values = [pokemon['spawn_chance'] for pokemon in data]
    plt.bar(labels, values)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(rotation=90)
    plt.show()


def main():
    filename = "pokemon_data.xlsx"


    structured_data = read_data_from_excel(filename)

    # Perform analysis

    spawn_rate_threshold = 0.05
    low_spawn_rate_pokemons = [pokemon for pokemon in structured_data if pokemon['spawn_chance'] < spawn_rate_threshold]
    print("Pokemons with spawn rate less than 5%:")
    print([pokemon['name'] for pokemon in low_spawn_rate_pokemons])
    print()

    # Plot for Pokemons with spawn rate less than 5%
    draw_bar_plot(low_spawn_rate_pokemons, 'Pokémon', 'Spawn Chance', 'Pokémons with Spawn Rate Less than 5%')

    max_weaknesses = 4
    pokemons_with_few_weaknesses = [pokemon for pokemon in structured_data if len(pokemon['weaknesses'].split(', ')) < max_weaknesses]
    print("Pokemons with less than 4 weaknesses:")
    print([pokemon['name'] for pokemon in pokemons_with_few_weaknesses])
    print()

    # Plot for Pokemons that have less than 4 weaknesses
    draw_bar_plot(pokemons_with_few_weaknesses, 'Pokémon', 'Weaknesses Count', 'Pokémons with Less than 4 Weaknesses')


    pokemons_without_multipliers = [pokemon for pokemon in structured_data if not pokemon['multipliers']]
    print("Pokemons without multipliers:")
    print([pokemon['name'] for pokemon in pokemons_without_multipliers])
    print()


    # Plot for Pokemons that have no multipliers at all
    draw_bar_plot(pokemons_without_multipliers, 'Pokémon', 'Has Multipliers', 'Pokémons without Multipliers')

    max_evolutions = 2
    pokemons_with_few_evolutions = [pokemon for pokemon in structured_data if len(pokemon['next_evolution']) <= max_evolutions]
    print("Pokemons with 2 or fewer evolutions:")
    print([pokemon['name'] for pokemon in pokemons_with_few_evolutions])
    print()
    # Plot for Pokemons that do not have more than 2 evolutions
    draw_bar_plot(pokemons_with_few_evolutions, 'Pokémon', 'Evolution Count', 'Pokémons with 2 or Fewer Evolutions')

    max_spawn_time = 300
    pokemons_with_low_spawn_time = [pokemon for pokemon in structured_data if isinstance(pokemon['spawn_time'], str) and int(pokemon['spawn_time'].split(':')[0]) * 60 + int(pokemon['spawn_time'].split(':')[1]) < max_spawn_time]
    print("Pokemons with spawn time less than 300 seconds:")
    print([pokemon['name'] for pokemon in pokemons_with_low_spawn_time])

    # Plot for Pokemons whose spawn time is less than 300 seconds
    draw_bar_plot(pokemons_with_low_spawn_time, 'Pokémon', 'Spawn Time (seconds)', 'Pokémons with Spawn Time Less than 300 seconds')

    pokemons_with_multiple_types = [pokemon for pokemon in structured_data if len(pokemon['type'].split(', ')) > 1]
    print("Pokemons with more than two types of capabilities:")
    print([pokemon['name'] for pokemon in pokemons_with_multiple_types])
    print()

    # Plot for Pokemon who have more than two types of capabilities
    draw_bar_plot(pokemons_with_multiple_types, 'Pokémon', 'Types Count', 'Pokémons with More than Two Types of Capabilities')

# Run the program
if __name__ == "__main__":
    main()
```
